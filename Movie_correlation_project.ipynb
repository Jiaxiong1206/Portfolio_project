```python
# Import libraries

import pandas as pd
import seaborn as sns 
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt 
plt.style.use('ggplot')
from matplotlib.pyplot import figure 

%matplotlib inline 
matplotlib.rcParams['figure.figsize'] = (12,8)  # Adjust the configuration of the plots created


# read the data
df = pd.read_csv(r'C:\Users\jiaxi\Downloads\movies.csv')
```

```python
# look at the data
df.head()
```

```python
# Look for any missing data 

for col in df.columns:
    missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,missing))
```

```python
# drop all the null rows 
df.dropna(inplace=True)
```

```python
for col in df.columns:
    missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,missing))
```

```python
# Data types for columns 
df.dtypes
```

```python
# convert data type to make the table cleaner 
df['gross'] = df['gross'].astype('int64')
df['budget'] = df['budget'].astype('int64')
```

```python
df.head()
```

```python
#create correct year column 

df['Correct year'] = df['released'].astype(str).str[-21:-15]
```

```python
df.head()
```

```python
# Check whether there are any non identical year with the correct year 

non_identical = (df['year'] != df['Correct year']).any()
print(non_identical)
```

```python
df.sort_values(by = ['gross'], inplace = False, ascending = False)
```

```python
pd.set_option('display.max_rows', 10)
```

```python
df.head()
```

```python
# To drop any duplicates

df['company'].drop_duplicates().sort_values(ascending = False).head()
```

```python
# Budget correlation with the gross
# Build a scatter plot with Budget vs gross revenue 

plt.scatter(x = df['budget'], y = df['gross'])
plt.title('Budget vs Gross Revenues')
plt.xlabel('Gross')
plt.ylabel('Budget revenues')
plt.show()
```

```python
# Plot Budget vs Gross Revenue with Seaborn

sns.regplot(x = 'budget', y = 'gross', data = df, scatter_kws = {"color": "red"}, line_kws = {"color" : "blue"})
plt.show()
```

```python
# Look at the correlation 
numerical_data = df.select_dtypes(include = ['float64','int64'])
numerical_data.corr(method = 'pearson' )  # Pearson, Kendall, Spearman
```

```python
# High correlation between budget and gross
```

```python
correlation_matrix = numerical_data.corr(method = 'pearson' )
sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation Matrix for Numeric Data')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# Look at company correlation with the gross

df_numerized = df
for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
df_numerized
        
```

```python
# see the correlation with a heatmap 

correlation_matrix = df_numerized.corr(method = 'pearson' )
sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation Matrix for Numeric Data')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values(ascending = False)
sorted_pairs
```

```python
high_corr = sorted_pairs [(sorted_pairs) > 0.5]
high_corr
```
